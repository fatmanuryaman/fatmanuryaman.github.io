```python
data_target = raw_data['target']
raw_data.drop(columns=['slope','target'],inplace=True)
```

```python
data_dummies=raw_data[['sex','cp','fbs','restecg','exang','ca','thal']]
data_dummies= pd.get_dummies(data_dummies,columns=['sex','cp','fbs','restecg','exang','ca','thal'])
```

```python
raw_data.drop(columns=['sex','cp','fbs','restecg','exang','ca','thal'],inplace=True)
data_final=raw_data.merge(data_dummies,left_index=True, right_index=True,how='left')
data_final.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data_final,data_target,test_size=0.3,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
from sklearn import metrics
lg_reg = LogisticRegression()
lg_reg.fit(x_train,y_train)
prediction = lg_reg.predict(x_test)
train_score = lg_reg.score(x_train,y_train)
test_score = lg_reg.score(x_test,y_test)
cm = confusion_matrix(y_test,prediction)
```

```python
print("Training score : "  , train_score)
print("Testing score : " , test_score)
```

```python
plot_confusion_matrix(lg_reg , x_test , y_test)
```
