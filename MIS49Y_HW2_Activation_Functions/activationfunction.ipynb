```python
import numpy as np
import matplotlib.pyplot as plt

#The linear activation function simply returns the input value.
#no activation
def linear(x):
   return x
#There is a certain step which is 0.
#Above the threshold 1, if below 0.
#0 for negative values, 1 for positive values.
def binary_step(x, threshold=0):
   return np.where(x > threshold, 1, 0)
#Swish is a sigmoid-like function.
#It returns x*sigmoid(x).
def swish(x, beta=1):
   return x * (1 / (1 + np.exp(-beta * x)))
#The sigmoid equation.
def sigmoid(x):
   return 1 / (1 + np.exp(-x))
#Softmax is the 
def softmax(x):
   exp_x = np.exp(x - np.max(x))
   return exp_x / exp_x.sum()
#The tanh function is already defined in numpy.
def tanh(x):
   return np.tanh(x)
#The ReLU returns max(0,x).
def relu(x):
   return np.maximum(0, x)
#The Leaky ReLU returns x*slope for negative values.
def leaky_relu(x, alpha=0.1):
   return np.where(x >= 0, x, alpha * x)
#The Parametrized ReLU changes the slope.
def parametric_relu(x, alpha=0.01):
   return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
#ELU creates exponential nonlinearity on negative inputs.
def elu(x, alpha=1.0):
   return np.where(x >= 0, x, alpha * (np.exp(x) - 1))
```

```python
# Define the input range
x = np.linspace(-10, 10, 500)

# Calculate the output values for each activation function for using them as y axis.
y_sigmoid = sigmoid(x)
y_softmax = softmax(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_leaky_relu = leaky_relu(x)
y_parametric_relu = parametric_relu(x)
y_elu = elu(x)
```

```python
# Create the subplots
fig, axes = plt.subplots(3, 3, figsize=(18, 12))


# Sigmoid activation function
axes[0, 0].plot(x, y_sigmoid)
axes[0, 0].set_title("Sigmoid Activation Function")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("y")


# Softmax activation function
axes[0, 1].plot(x, y_softmax)
axes[0, 1].set_title("Softmax Activation Function")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("y")


# Tanh activation function
axes[0, 2].plot(x, y_tanh)
axes[0, 2].set_title("Tanh Activation Function")
axes[0, 2].set_xlabel("x")
axes[0, 2].set_ylabel("y")


# ReLU activation function
axes[1, 0].plot(x, y_relu)
axes[1, 0].set_title("ReLU Activation Function")
axes[1, 0].set_xlabel("x")
axes[1, 0].set_ylabel("y")


# Leaky ReLU activation function
axes[1, 1].plot(x, y_leaky_relu)
axes[1, 1].set_title("Leaky ReLU Activation Function")
axes[1, 1].set_xlabel("x")
axes[1, 1].set_ylabel("y")


# Parameterized ReLU activation function
axes[1, 2].plot(x, y_parametric_relu)
axes[1, 2].set_title("Parameterized ReLU Activation Function")
axes[1, 2].set_xlabel("x")
axes[1, 2].set_ylabel("y")


# ELU activation function
axes[2, 0].plot(x, y_elu)
axes[2, 0].set_title("ELU Activation Function")
axes[2, 0].set_xlabel("x")
axes[2, 0].set_ylabel("y")
axes[2, 1].axis("off")
axes[2, 2].axis("off")
```
